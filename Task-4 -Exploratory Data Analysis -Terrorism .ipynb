# Terrorist Activities over the years 

```python
df1 = df.groupby('iyear')['eventid'].count().sort_values(ascending=False).to_frame()

df1.head()
```

```python
sns.set()
plt.figure(figsize=(15,8))
sns.lineplot(df1.index,df1['eventid'])
plt.xlabel('Year')
plt.ylabel('Count of Terrorist Attacks Reported')
plt.show()
```

## We can see there is an upward trend of Terrorist activities over the years staring from 1970 till now. It was highest in the year 2014.

```python
df.head()
```

# Terrorist Activities Country wise statistics

```python
df2 = df.groupby('country_txt')['eventid'].count().sort_values(ascending=False).to_frame()

df2 = df2.head(20)

df2.plot(kind='barh',figsize=(20,15),legend=False)

plt.ylabel('Top 20 countries with highest amount of Terrorist Attacks')

plt.xlabel('Count')
```

#### Conclusion : Iraq ,Pakistan and Afganisthan are Top 3 countries with Highest number of Terrorist Attacks.

# Terrorist Activities Region wise statistics

```python
df2 = df.groupby(['region_txt'])['eventid'].count().sort_values(ascending=False).to_frame()

df2.plot(kind='barh',figsize=(20,15),legend=False,colormap='Paired_r')

plt.ylabel('Top 12 regions with their regions with highest amount of Terrorist Attacks')

plt.xlabel('Count')

plt.show()
```

### Conclusion : Middle East-North Afirca, South Asia and South-America are Top 3 regions with Highest number of Terrorist Attacks.

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
crs = {'init':'epsg:4326'}
```

```python
geometry = [Point(xy) for xy in zip(df['longitude'],df['latitude'])]
```

```python
geometry[:10]
```

```python
geo_df = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
```

```python
geo_df.iloc[:,-1:].head()
```

## Checking the Most Frequent Type of Terrorist Attack

```python
plt.figure(figsize=(15,8))
sns.countplot(y=geo_df['attacktype1_txt'], order = geo_df['attacktype1_txt'].value_counts().index)
plt.ylabel('Attack Type')
```

### Conclusion : We can see Bombing/Explosion , Armed Assault and Assassination are the most frequent type of Attack.

# Geo-locations of Most Frequent Type of Terrorist Attack


```python
geo_df['attacktype1_txt'].value_counts()
```

```python
geo_df['attacktype1_txt'].unique()
```

```python
country_map= gpd.read_file('World_Countries.shp')

fig,ax = plt.subplots(figsize=(20,20))
country_map.plot(ax=ax, alpha=0.5, color='grey')

geo_df[geo_df['attacktype1_txt']=='Bombing/Explosion']['geometry'].plot(ax=ax, markersize=5,color='red',marker='o',label='Bombing/Explosion')

geo_df[geo_df['attacktype1_txt']=='Armed Assault']['geometry'].sample(n=18000,random_state=0).plot(ax=ax, markersize=5,color='blue',marker='v',label='Armed Assault')

geo_df[geo_df['attacktype1_txt']=='Hostage Taking (Kidnapping)']['geometry'].plot(ax=ax, markersize=5,color='green',marker='^',label='Hostage Taking(Kidnapping)')


plt.legend(prop={'size':20})
```

## Checking the Most Frequent Target of Terrorist Attack

```python
plt.figure(figsize=(15,20))
sns.countplot(y=geo_df['targtype1_txt'], order = geo_df['targtype1_txt'].value_counts().index)
plt.ylabel('Target Type')
```

### Conclusion : We can see Private Citizens & Property , Military and Police are the most frequent target of Attack.
